```python
# Program 66: Tables 1–10
def main():
    for i in range(1, 11):
        print("Table of", i)
        
        for j in range(1, 6):
            print(i, "x", j, "=", i*j)

main()
```

```python
# Program 67: Star Pattern
def main():
    n = int(input("Enter rows: "))
    
    for i in range(1, n+1):
        for j in range(i):
            print("*", end="")
        print()

main()
```

```python
# Program 68: Sum Even
def main():
    n = int(input("Enter limit: "))
    
    total = 0
    
    for i in range(1, n+1):
        if i % 2 == 0:
            total += i
    
    print("Sum of even:", total)

main()
```
