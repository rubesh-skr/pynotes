```python
# Program 42: Access Value
def main():
    data = {"a": 10, "b": 20}
    
    key = input("Enter key: ")
    
    if key in data:
        print("Value:", data[key])
    else:
        print("Key not found")

main()
```

```python
# Program 43: Update Dictionary
def main():
    data = {"a": 10, "b": 20}
    
    key = input("Enter key: ")
    val = int(input("Enter value: "))
    
    data[key] = val
    
    print("Updated:", data)

main()
```

```python
# Program 44: Delete Key
def main():
    data = {"a": 10, "b": 20}
    
    key = input("Enter key: ")
    
    if key in data:
        del data[key]
    
    print("Result:", data)

main()
```
