```python
# Program 36: Sort List
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    
    nums.sort()
    
    print("Sorted:", nums)

main()
```

```python
# Program 37: Second Largest
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    
    nums = list(set(nums))
    nums.sort()
    
    print("Second largest:", nums[-2])

main()
```

```python
# Program 38: Merge Lists
def main():
    a = list(map(int, input("Enter list1: ").split()))
    b = list(map(int, input("Enter list2: ").split()))
    
    merged = a + b
    
    print("Merged list:", merged)

main()
```
