```python
# Program 30: Sum of List
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    
    total = 0
    
    for n in nums:
        total = total + n
    
    print("Sum is:", total)

main()
```

```python
# Program 31: Average
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    
    total = 0
    
    for n in nums:
        total += n
    
    avg = total / len(nums)
    
    print("Average:", avg)

main()
```

```python
# Program 32: Count Even
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    
    count = 0
    
    for n in nums:
        if n % 2 == 0:
            count += 1
    
    print("Even count:", count)

main()
```
