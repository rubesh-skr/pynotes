```python
# Program 57: Class
class Person:
    def show(self):
        print("Hello")

def main():
    p = Person()
    p.show()

main()
```

```python
# Program 58: Constructor
class Person:
    def __init__(self, name):
        self.name = name
    
    def show(self):
        print("Name:", self.name)

def main():
    p = Person("Ram")
    p.show()

main()
```

```python
# Program 59: Inheritance
class A:
    def show(self):
        print("Class A")

class B(A):
    pass

def main():
    obj = B()
    obj.show()

main()
```
