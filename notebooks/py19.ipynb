```python
# Program 51: Write File
def main():
    f = open("test.txt", "w")
    
    f.write("Hello Python\n")
    f.write("File Writing\n")
    
    f.close()
    
    print("File written")

main()
```

```python
# Program 52: Read File
def main():
    f = open("test.txt", "r")
    
    content = f.read()
    
    print("Content:")
    print(content)
    
    f.close()

main()
```

```python
# Program 53: Count Lines
def main():
    f = open("test.txt", "r")
    
    lines = f.readlines()
    
    count = len(lines)
    
    print("Total lines:", count)
    
    f.close()

main()
```
