```python
numbers = [10, 20, 30, 40]

total = 0
for num in numbers:
    total += num   # shorter and clean

print("Total:", total)
```

```python
numbers = [1, 2, 3, 4, 5, 6]

count = 0
for num in numbers:
    if num % 2 == 0:
        count += 1

print("Even Count:", count)
```

```python
numbers = [5, 25, 10, 15]

largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest:", largest)
```
