```python
def count_words():
    text = input("Enter sentence: ")
    words = text.split()

    count = 0
    for w in words:
        count += 1

    print("Word count:", count)

while True:
    print("\n1.Count 2.Again 3.Skip 4.Exit")
    ch = input("Choice: ")

    if ch == "1" or ch == "2":
        count_words()
    elif ch == "3":
        print("Skip")
    elif ch == "4":
        print("Program exited")
        break
    else:
        print("Invalid choice")
```
