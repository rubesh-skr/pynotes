```python
# Program 27: Count Words
def main():
    text = input("Enter sentence: ")
    
    words = text.split()
    
    count = 0
    
    for w in words:
        count += 1
    
    print("Total words:", count)

main()
```

```python
# Program 28: Largest in List
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    
    largest = nums[0]
    
    for n in nums:
        if n > largest:
            largest = n
    
    print("Largest:", largest)

main()
```

```python
# Program 29: Smallest in List
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    
    smallest = nums[0]
    
    for n in nums:
        if n < smallest:
            smallest = n
    
    print("Smallest:", smallest)

main()
```
