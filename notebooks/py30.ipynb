```python
# Program 84: Remove Element
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    val = int(input("Enter value: "))
    
    result = []
    
    for n in nums:
        if n != val:
            result.append(n)
    
    print("Result:", result)

main()
```

```python
# Program 85: Count Lines
def main():
    print("Enter lines (type stop to end):")
    
    count = 0
    
    while True:
        line = input()
        if line == "stop":
            break
        count += 1
    
    print("Total lines:", count)

main()
```

```python
# Program 86: Min Digit
def main():
    num = int(input("Enter number: "))
    
    min_d = 9
    
    while num > 0:
        d = num % 10
        if d < min_d:
            min_d = d
        num = num // 10
    
    print("Min digit:", min_d)

main()
```
