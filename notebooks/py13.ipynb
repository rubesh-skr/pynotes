```python
# Program 33: Count Odd
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    
    count = 0
    
    for n in nums:
        if n % 2 != 0:
            count += 1
    
    print("Odd count:", count)

main()
```

```python
# Program 34: Reverse List
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    
    rev = nums[::-1]
    
    print("Reversed:", rev)

main()
```

```python
# Program 35: Remove Duplicates
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    
    unique = []
    
    for n in nums:
        if n not in unique:
            unique.append(n)
    
    print("Unique list:", unique)

main()
```
