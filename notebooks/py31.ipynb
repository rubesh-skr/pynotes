```python
# Program 87: Reverse Words
def main():
    text = input("Enter sentence: ")
    
    words = text.split()
    
    rev = words[::-1]
    
    print("Reversed words:", " ".join(rev))

main()
```

```python
# Program 88: Capitalize
def main():
    text = input("Enter sentence: ")
    
    words = text.split()
    
    result = []
    
    for w in words:
        result.append(w.capitalize())
    
    print("Result:", " ".join(result))

main()
```

```python
# Program 89: Substring
def main():
    text = input("Enter text: ")
    sub = input("Enter substring: ")
    
    if sub in text:
        print("Found")
    else:
        print("Not found")

main()
```
