```python
# Program 45: Dictionary Length
def main():
    data = {"a": 1, "b": 2, "c": 3}
    
    count = 0
    
    for k in data:
        count += 1
    
    print("Length:", count)

main()
```

```python
# Program 46: Loop Dictionary
def main():
    data = {"a": 10, "b": 20}
    
    print("Items:")
    
    for k in data:
        print(k, data[k])

main()
```

```python
# Program 47: Check Key
def main():
    data = {"x": 5, "y": 10}
    
    key = input("Enter key: ")
    
    if key in data:
        print("Exists")
    else:
        print("Not exists")

main()
```
