```python
# Program 2: Addition
def main():
    a = 10
    b = 20
    
    print("First number:", a)
    print("Second number:", b)
    
    result = a + b
    
    print("Sum is:", result)

main()
```

```python
# Program 3: Addition using input
def main():
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    
    print("You entered:", a, b)
    
    result = a + b
    
    print("Total:", result)

main()
```

```python
# Program 4: Even or Odd
def main():
    num = int(input("Enter number: "))
    
    print("Checking number...")
    
    if num % 2 == 0:
        print("Even number")
    else:
        print("Odd number")

main()
```
