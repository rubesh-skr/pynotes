```python
# Program 54: Copy File
def main():
    f1 = open("test.txt", "r")
    f2 = open("copy.txt", "w")
    
    data = f1.read()
    
    f2.write(data)
    
    f1.close()
    f2.close()

main()
```

```python
# Program 55: Exception
def main():
    try:
        a = int(input("Enter number: "))
        b = int(input("Enter number: "))
        
        print(a / b)
    
    except:
        print("Error occurred")

main()
```

```python
# Program 56: Custom Exception
def main():
    try:
        age = int(input("Enter age: "))
        
        if age < 18:
            raise Exception("Too young")
        
        print("Allowed")
    
    except Exception as e:
        print(e)

main()
```
