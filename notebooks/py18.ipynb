```python
# Program 48: Sum Values
def main():
    data = {"a": 10, "b": 20, "c": 30}
    
    total = 0
    
    for v in data.values():
        total += v
    
    print("Sum:", total)

main()
```

```python
# Program 49: Copy Dictionary
def main():
    data = {"a": 1, "b": 2}
    
    copy_data = {}
    
    for k in data:
        copy_data[k] = data[k]
    
    print("Copy:", copy_data)

main()
```

```python
# Program 50: Merge Dict
def main():
    d1 = {"a": 1}
    d2 = {"b": 2}
    
    merged = {}
    
    for k in d1:
        merged[k] = d1[k]
    
    for k in d2:
        merged[k] = d2[k]
    
    print("Merged:", merged)

main()
```
