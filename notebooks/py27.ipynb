```python
# Program 75: Find Index
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    target = int(input("Enter element: "))
    
    for i in range(len(nums)):
        if nums[i] == target:
            print("Index:", i)
            break

main()
```

```python
# Program 76: Count Occurrence
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    target = int(input("Enter element: "))
    
    count = 0
    
    for n in nums:
        if n == target:
            count += 1
    
    print("Count:", count)

main()
```

```python
# Program 77: Replace Element
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    old = int(input("Old value: "))
    new = int(input("New value: "))
    
    for i in range(len(nums)):
        if nums[i] == old:
            nums[i] = new
    
    print("Updated list:", nums)

main()
```
