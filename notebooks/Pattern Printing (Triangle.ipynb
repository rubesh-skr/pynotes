```python
username = "admin"
password = "1234"

def login():
    u = input("Enter username: ")
    p = input("Enter password: ")

    if u == username and p == password:
        print("Login Successful")
    else:
        print("Invalid credentials")

def attempts():
    tries = 3

    while tries > 0:
        login()
        tries -= 1
        print("Attempts left:", tries)

attempts()
```

```python
def sum_digits():
    num = int(input("Enter number: "))
    total = 0

    while num > 0:
        digit = num % 10
        total += digit
        num = num // 10

    print("Sum of digits:", total)

def repeat():
    n = int(input("How many numbers: "))
    for i in range(n):
        sum_digits()

repeat()
```

```python
def pattern():
    n = int(input("Enter rows: "))

    i = 1
    while i <= n:
        j = 1
        while j <= i:
            print("*", end=" ")
            j += 1
        print()
        i += 1

def repeat():
    t = int(input("How many patterns: "))
    for i in range(t):
        pattern()

repeat()
```
