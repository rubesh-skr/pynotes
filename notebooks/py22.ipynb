```python
# Program 60: Override
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")

def main():
    obj = B()
    obj.show()

main()
```

```python
# Program 61: Calculator
def main():
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    op = input("Enter operator (+,-,*,/): ")
    
    if op == '+':
        print("Result:", a + b)
    elif op == '-':
        print("Result:", a - b)
    elif op == '*':
        print("Result:", a * b)
    elif op == '/':
        print("Result:", a / b)

main()
```

```python
# Program 62: Guess Number
def main():
    secret = 5
    
    guess = int(input("Guess number (1-10): "))
    
    if guess == secret:
        print("Correct!")
    else:
        print("Wrong! Answer is", secret)

main()
```
