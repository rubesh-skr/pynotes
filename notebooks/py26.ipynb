```python
# Program 72: Remove Spaces
def main():
    text = input("Enter text: ")
    
    result = ""
    
    for ch in text:
        if ch != " ":
            result += ch
    
    print("Without spaces:", result)

main()
```

```python
# Program 73: Uppercase Count
def main():
    text = input("Enter text: ")
    
    count = 0
    
    for ch in text:
        if ch.isupper():
            count += 1
    
    print("Uppercase letters:", count)

main()
```

```python
# Program 74: Lowercase Count
def main():
    text = input("Enter text: ")
    
    count = 0
    
    for ch in text:
        if ch.islower():
            count += 1
    
    print("Lowercase letters:", count)

main()
```
