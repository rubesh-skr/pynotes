```python
# Program 78: Check Sorted
def main():
    nums = list(map(int, input("Enter numbers: ").split()))
    
    if nums == sorted(nums):
        print("Sorted")
    else:
        print("Not sorted")

main()
```

```python
# Program 79: Matrix Sum
def main():
    m = [[1,2],[3,4]]
    
    total = 0
    
    for row in m:
        for val in row:
            total += val
    
    print("Sum:", total)

main()
```

```python
# Program 80: Diagonal Sum
def main():
    m = [[1,2],[3,4]]
    
    total = 0
    
    for i in range(len(m)):
        total += m[i][i]
    
    print("Diagonal sum:", total)

main()
```
