```python
# Program 69: Sum Odd
def main():
    n = int(input("Enter limit: "))
    
    total = 0
    
    for i in range(1, n+1):
        if i % 2 != 0:
            total += i
    
    print("Sum of odd:", total)

main()
```

```python
# Program 70: Max Digit
def main():
    num = int(input("Enter number: "))
    
    max_d = 0
    
    while num > 0:
        d = num % 10
        if d > max_d:
            max_d = d
        num = num // 10
    
    print("Max digit:", max_d)

main()
```

```python
# Program 71: Count Spaces
def main():
    text = input("Enter text: ")
    
    count = 0
    
    for ch in text:
        if ch == " ":
            count += 1
    
    print("Spaces:", count)

main()
```
